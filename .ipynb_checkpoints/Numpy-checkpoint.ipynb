```python
# how to create an array using arrange function
new_array=np.arange((4))
print(new_array)
'''
To create an array using NumPy's arange() function, 
you can think of it like Python's built-in range() â€” but it returns a NumPy array instead of a list.
np.arange(start, stop, step)
- start: starting value (inclusive)
- stop: ending value (exclusive)
- step: increment (default is 1)
'''
```

```python
# You can create a diagonal array using NumPyâ€™s eye() function, 
# which generates a 2D array with ones on the main diagonal and zeros elsewhere.
# np.eye(N, M=None, k=0)
# - N: number of rows
# - M: number of columns (optional; defaults to N)
# - k: index of the diagonal
# - k=0: main diagonal
# - k>0: above the main diagonal
# - k<0: below the main diagonal
new_array=np.eye(3)
print(new_array)

```
