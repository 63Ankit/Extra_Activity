```python
import numpy as np
import matplotlib.pyplot as plt

# Runs data
sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 200, 900, 1400, 1600, 1800, 1500, 1100, 800, 0]
kohli  = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]

# Define years
years = np.arange(2001, 2011)  # Example: 20012010

y = np.arange(len(years))  # index positions
height = 0.25              # thickness of each bar

# Horizontal bars
plt.barh(y - height, sachin, height=height, label="Sachin")
plt.barh(y, sehwag, height=height, label="Sehwag")
plt.barh(y + height, kohli, height=height, label="Kohli")

# Labels and title
plt.ylabel("Year")
plt.xlabel("Runs")
plt.title("Run Comparison (Horizontal)")
plt.yticks(y, years)  # Show actual years instead of 0,1,2,...
plt.legend()
plt.tight_layout()
plt.show()
```
